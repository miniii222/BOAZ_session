```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://comic.naver.com/webtoon/weekday.nhn"
response = requests.get(url)
html = response.text
```

```python
soup = BeautifulSoup(html, "lxml")
webtoon = soup.select("a.title")
for w in webtoon :
    print(w.text)
```

```python
day = soup.select_one("h4.tue span")
print(day.text)
```

```python
days = soup.select("h4 span")
for d in days :
    print(d.text)
```

```python
daylist = ["mon","tue","wed","thu","fri","sat","sun"]
w = soup.select("a.title")
for d in daylist :
    print(d)
    for wt in w :
        if wt.attrs["href"].split("=")[2] == d:
            print(wt.text)
```
