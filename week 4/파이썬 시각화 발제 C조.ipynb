## 그림범위 &틱설정 & 그리드 설정

### x,y의 범위 : plt.xlim(a,b),plt.ylim(c,d) // x,y의 틱 : plt.xticks(), plt.yticks()

### 그리드설정 :  plt.grid(True)

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b", 
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")

plt.xlim(0, 50)
plt.ylim(-10, 30)

plt.show()
```
